# Libraries

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import re
from nltk.tokenize import sent_tokenize, word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer

from math import ceil
```

```python
from tqdm import tqdm
```

# Data

```python
text = pd.read_csv('InshortsScraped.csv')['news']
```

```python
text.head()
```

```python
text.shape
```

# EDA

```python
with open('stopwords.txt') as f:
    stopwords_ = f.readlines()
```

```python
stopwords = set(map(lambda x: x.strip(), stopwords_))|set(['.'])
```

```python
print(sorted(stopwords))
```

```python
def remove_stop_words(X):
    features = []
    for x in X.split():
        if x.lower() not in stopwords and len(x) > 2:
            features.append(x.lower())
    return ' '.join(features)
```

```python
text = text.apply(lambda x : re.sub(r'\W', ' ', str(x)))
text = text.apply(lambda x : re.sub(r'\d', ' ', str(x)))
text = text.apply(lambda x : re.sub(r'\.', ' . ', str(x)))
text = text.apply(lambda x : re.sub(r'\s+', ' ', str(x)))
text = text.apply(lambda x : x.strip())
text = text.apply(remove_stop_words)
```

```python
fig, ax = plt.subplots(figsize=(25,8))
pd.Series(' '.join(text).split()).value_counts()[:20].plot.barh(ax = ax, fontsize=12)
ax.set_title(f'Most Frequent Words')
fig.gca().invert_yaxis()
```

# Vectorization

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(text)
vectors = vectorizer.transform(text)
```

```python
vectors.shape
```

# Clustering

## K Means

```python
from sklearn.cluster import KMeans
```

```python
n = 5
model = KMeans(n_clusters=n)
model.fit(vectors)
print('Clustering Done !!!')
y_pred = pd.Series(model.predict(vectors), name='Cluster')
vals = y_pred.value_counts()
vals.plot.pie(figsize=(6, 6), labels=vals)
plt.legend(vals.index)
```

```python
figsize = (25, 10*ceil(n/2))
fig, ax = plt.subplots(figsize=figsize, ncols=2, nrows=ceil(n/2))
for i in range(n):
    pd.Series(' '.join(text[y_pred == i]).split()).value_counts()[::-1].tail(20).plot.barh(ax = ax[i//2][i%2], fontsize=16)
    ax[i//2][i%2].set_title(f'Most Frequent Words in cluster {i}')
fig.gca().invert_yaxis()
```

## Spectral Clustering

```python
from sklearn.cluster import SpectralClustering
```

```python
n = 5
model = SpectralClustering(n_clusters=n)
y_pred = pd.Series(model.fit_predict(vectors), name='Cluster')
vals = y_pred.value_counts()
vals.plot.pie(figsize=(6, 6), labels=vals)
plt.legend(vals.index)
```

```python
figsize = (25, 10*ceil(n/2))
fig, ax = plt.subplots(figsize=figsize, ncols=2, nrows=ceil(n/2))
for i in range(n):
    pd.Series(' '.join(text[y_pred == i]).split()).value_counts()[::-1].tail(20).plot.barh(ax = ax[i//2][i%2], fontsize=16)
    ax[i//2][i%2].set_title(f'Most Frequent Words in cluster {i}')
fig.gca().invert_yaxis()
```

# Word2Vec

## CBOW

```python
from gensim.models import word2vec
```

```python
words = []
for t in text:
    words.append(t.split())

w2v = word2vec.Word2Vec(words, size=100)
```

```python
words
```

## Multiply with TFIDF Weights

```python
vectors_ = []
for sent in tqdm(text):
    v = np.zeros((100,))
    for word in sent.split():
        try:
            v += w2v.wv[word]*vectorizer.idf_[vectorizer.vocabulary_[word]]
        except Exception as e:
#             print(e)
            pass
    vectors_.append(v)
```

# Clustering

## K Means

```python
n = 5
model = KMeans(n_clusters=n)
model.fit(vectors_)
print('Clustering Done !!!')
y_pred = pd.Series(model.predict(vectors_), name='Cluster')
vals = y_pred.value_counts()
vals.plot.pie(figsize=(6, 6), labels=vals)
plt.legend(vals.index)
```

```python
figsize = (25, 10*ceil(n/2))
fig, ax = plt.subplots(figsize=figsize, ncols=2, nrows=ceil(n/2))
for i in range(n):
    pd.Series(' '.join(text[y_pred == i]).split()).value_counts()[::-1].tail(20).plot.barh(ax = ax[i//2][i%2], fontsize=16)
    ax[i//2][i%2].set_title(f'Most Frequent Words in cluster {i}')
fig.gca().invert_yaxis()
```

# Word Clustering

```python
words_ = list(w2v.wv.vocab.keys())
```

```python
vect_df = pd.DataFrame(columns = ['x'+str(i) for i in range(1, 101)])
```

```python
for i, w in tqdm(enumerate(words_)):
    vect_df.loc[i] = w2v.wv[w]
```

```python
n = 5
model = KMeans(n_clusters=n)
model.fit(vect_df)
print('Clustering Done !!!')
y_pred = pd.Series(model.predict(vect_df), name='Cluster')
vals = y_pred.value_counts()
vals.plot.pie(figsize=(6, 6), labels=vals)
plt.legend(vals.index)
```

```python
# figsize = (25, 10*ceil(n/2))
# fig, ax = plt.subplots(figsize=figsize, ncols=2, nrows=ceil(n/2))
clusters = [[],[],[],[],[]]
for i in range(n):
    for j in range(len(y_pred)):
        if y_pred[j] == i:
            clusters[i].append(words_[j])
#     pd.Series(' '.join(words_[y_pred == i]).split()).value_counts()[::-1].tail(20).plot.barh(ax = ax[i//2][i%2], fontsize=16)
#     ax[i//2][i%2].set_title(f'Most Frequent Words in cluster {i}')
# fig.gca().invert_yaxis()
```
